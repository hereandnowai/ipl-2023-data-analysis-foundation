# Section 6: PROJECT-SPECIFIC CONCEPTS - Linear Regression Math

Linear Regression is finding the "Line of Best Fit."

The formula for a line is:
$$y = mx + b$$

In AI, we usually write it as:
$$\hat{y} = w \cdot x + b$$

Where:
- $\hat{y}$ is the predicted price.
- $w$ is the **weight** (how much each room adds to the price).
- $x$ is the number of rooms.
- $b$ is the **bias** (starting price).

### Simple Implementation with NumPy
We can find the best $w$ and $b$ using the **Mean Squared Error (MSE)** and **Gradient Descent**, but for this project, we will use the **Simple Statistics Approach**:
$$w = \frac{\text{Covariance}(x, y)}{\text{Variance}(x)}$$
$$b = \bar{y} - w \cdot \bar{x}$$
