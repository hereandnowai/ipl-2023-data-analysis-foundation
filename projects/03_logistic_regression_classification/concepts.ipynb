# Section 6: PROJECT-SPECIFIC CONCEPTS - Logistic Regression Logic

Logistic Regression uses the **Sigmoid Function** to squash linear predictions into a range between 0 and 1.

$$P(y=1) = \frac{1}{1 + e^{-(wx + b)}}$$

### Thresholding
If $P > 0.5$, we predict Class 1 (Diabetes).
If $P \leq 0.5$, we predict Class 0 (No Diabetes).


```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Visualizing the Sigmoid Curve
z = np.linspace(-10, 10, 100)
p = sigmoid(z)

plt.plot(z, p)
plt.axhline(0.5, color='red', linestyle='--')
plt.title('Sigmoid Function (The basis of Classification)')
plt.xlabel('z (wx + b)')
plt.ylabel('Probability')
plt.show()
```
