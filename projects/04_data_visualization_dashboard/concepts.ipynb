# Section 5: MATPLOTLIB BASICS - Subplots

Subplots allow you to put multiple charts in one figure.

### Code Example
```python
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].plot([1, 2, 3], [4, 5, 6])
ax[1].bar(['A', 'B'], [10, 20])
plt.show()
```
