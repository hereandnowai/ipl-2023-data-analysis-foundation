# Movie Review Sentiment Analysis (Manual NLP)

## SECTION 1: Problem Statement
Classify IMDB reviews as **Positive** or **Negative** using a simple Keyword Matching algorithm.


```python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv('imdb_reviews.csv')
# Look at first few reviews
df.head()

# Simple Sentiment Keyword Lists
pos_words = set(['great', 'excellent', 'amazing', 'brilliant', 'wonderful', 'perfect', 'best'])
neg_words = set(['bad', 'awful', 'terrible', 'worst', 'poor', 'boring', 'disappointed', 'waste'])

def manual_sentiment(text):
    text = str(text).lower()
    words = text.split()
    score = 0
    for word in words:
        if word in pos_words: score += 1
        if word in neg_words: score -= 1
    return 1 if score > 0 else 0

# Test individual review
review = df.iloc[0]['review']
print(f"Sample Review: {review[:100]}...")
print(f"Calculated Sentiment: {manual_sentiment(review)}")
```
