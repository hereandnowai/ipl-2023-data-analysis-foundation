# Estimating $\pi$ (Monte Carlo Simulation)

## SECTION 1: Problem Statement
Estimate the value of $\pi$ by simulating random "dart throws" at a 1x1 square containing a circle of radius 0.5.


```python
import numpy as np
import matplotlib.pyplot as plt

# Number of "darts" to throw
N = 10000

# Random x, y coords between -1 and 1
X = np.random.uniform(-1, 1, N)
Y = np.random.uniform(-1, 1, N)

# Check which hits the circle x^2 + y^2 <= 1
distance = X**2 + Y**2
inside_circle = distance <= 1

# Est Pi = 4 * (hits / total)
pi_estimate = 4 * np.sum(inside_circle) / N
print(f"Estimated Pi (N={N}): {pi_estimate}")

# Visualization
plt.figure(figsize=(8, 8))
plt.scatter(X[inside_circle], Y[inside_circle], color='blue', s=1, label='Inside')
plt.scatter(X[~inside_circle], Y[~inside_circle], color='red', s=1, label='Outside')
plt.title(f'Monte Carlo Pi Estimation (Pi ~ {pi_estimate})')
plt.legend()
plt.show()
```
