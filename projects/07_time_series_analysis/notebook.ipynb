# Stock Price Analysis (Time Series)

## SECTION 1: Problem Statement
Track Apple (AAPL) stock price over time and identify trends using Rolling Averages.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('apple_stock.csv')
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
df.sort_index(inplace=True)

# Calculate Rolling Average (30-day window)
df['Rolling_30'] = df['Close'].rolling(window=30).mean()

# Visualization
plt.figure(figsize=(14, 7))
plt.plot(df.index, df['Close'], label='Daily Close', alpha=0.3)
plt.plot(df.index, df['Rolling_30'], label='30-Day Rolling Mean', color='orange')
plt.title('Apple Stock Price (Close) and Rolling Average')
plt.xlabel('Year')
plt.ylabel('Price ($)')
plt.legend()
plt.show()
```
