# Section 3: NumPy Basics - Working with Numerical Arrays

### What is NumPy?
NumPy is short for **Numerical Python**. It's the foundation of almost all data science and AI tools.

### Real-World Analogy
Think of a NumPy array like a high-speed, mathematical spreadsheet. It can handle millions of calculations in a fraction of a second.


```python
import numpy as np

# Creating an array (like a row in a spreadsheet)
ages = np.array([22, 38, 26, 35, 33])

print("Array:", ages)
print("Shape (How many elements?):", ages.shape)
print("First Age:", ages[0])

# Math operations (Vectorization)
# In standard Python, adding 1 to all would require a loop.
# In NumPy, it's a single operation:
ages_plus_one = ages + 1
print("Ages next year:", ages_plus_one)
```

# Section 4: Pandas Basics - Data Manipulation with DataFrames

### What is Pandas?
If NumPy is for math, **Pandas** is for tables. The main structure in Pandas is the **DataFrame**.

### Real-World Analogy
A Pandas DataFrame is like **Excel on steroids**. You can sort, filter, and combine data much faster using code.


```python
import pandas as pd

# Creating a mini DataFrame
data = {
    'Name': ['Braund', 'Cumings', 'Heikkinen'],
    'Age': [22, 38, 26],
    'Survived': [0, 1, 1]
}
df = pd.DataFrame(data)

print("Full DataFrame:")
print(df)

# Basic exploration commands
print("\nFirst 2 rows:")
print(df.head(2))

print("\nSummary statistics:")
print(df.describe())
```

# Section 5: Matplotlib Basics - Data Visualization

### Why Visualize?
Charts make it easier to see patterns that are hidden in rows of numbers.

### Real-World Analogy
Visualizing data is like "telling a story" with your numbers.


```python
import matplotlib.pyplot as plt

# Comparison: Who survived?
categories = ['Died', 'Survived']
counts = [549, 342] # Sample numbers

plt.bar(categories, counts, color=['red', 'green'])
plt.xlabel('Outcome')
plt.ylabel('Count')
plt.title('Titanic Survival Overview')
plt.show()
```

# Section 6: Project-Specific Concepts - Data Cleaning Logic

### Filling the "Holes" (Null Values)
Some passengers didn't have their Age recorded. We can't just leave a "blank" when doing math.

Common strategies:
1. **isna()**: Identify where the holes are.
2. **fillna()**: Fill holes with a value (like the average age).
3. **drop()**: Remove useless information entirely.
