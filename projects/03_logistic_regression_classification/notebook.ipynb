# Logistic Regression from Scratch (Binary Classification)

## SECTION 1: Problem Statement
Predict whether a patient has diabetes based on diagnostic measurements.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load dataset (Pima Indians Diabetes)
column_names = ['preg', 'glucose', 'bp', 'skin', 'insulin', 'bmi', 'pedigree', 'age', 'target']
df = pd.read_csv('diabetes.csv', names=column_names)
df.head()
```

## SECTION 4: Building the Model (Hand-coded Logistic Regression)
We'll implement the Sigmoid and use a simple heuristic for weights.


```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Simple Classification using Glucose as the primary indicator
glucoses = df['glucose'].values
targets = df['target'].values

# Standardize glucose (Scale it between 0 and 1 roughly)
x_std = (glucoses - glucoses.mean()) / glucoses.std()

# Predictions (Simplified weights)
w = 1.0 # arbitrary start
b = 0.0 # arbitrary start
prob = sigmoid(w * x_std + b)
preds = (prob > 0.5).astype(int)

# Check Accuracy
accuracy = np.mean(preds == targets)
print(f"Accuracy with current weights: {accuracy * 100:.2f}%")
```
