# Supermarket Sales Dashboard

## SECTION 1: Problem Statement
Identify which product lines are the most profitable and how different payment methods are used across branches.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('supermarket_sales.csv')
df.head()
```

## SECTION 6: Creating the 2x2 Dashboard
We'll combine multiple visualizations into one report.


```python
fig, ax = plt.subplots(2, 2, figsize=(15, 12))

# 1. Sales by Branch
branch_sales = df.groupby('Branch')['Total'].sum()
ax[0, 0].bar(branch_sales.index, branch_sales.values, color='skyblue')
ax[0, 0].set_title('Total Sales by Branch')

# 2. Income by Product Line
product_income = df.groupby('Product line')['gross income'].sum()
ax[0, 1].barh(product_income.index, product_income.values, color='lightgreen')
ax[0, 1].set_title('Gross Income by Product Line')

# 3. Payment Distribution
payment_counts = df['Payment'].value_counts()
ax[1, 0].pie(payment_counts.values, labels=payment_counts.index, autopct='%1.1f%%')
ax[1, 0].set_title('Payment Method Distribution')

# 4. Quantity Distribution
ax[1, 1].hist(df['Quantity'], bins=10, color='orange', edgecolor='black')
ax[1, 1].set_title('Quantity per Transaction')

plt.tight_layout()
plt.show()
```
