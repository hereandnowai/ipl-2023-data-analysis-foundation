# Movie Recommendation System

## SECTION 1: Problem Statement
Build a simple engine that recommends movies based on **Genre Similarity**.


```python
import pandas as pd
import numpy as np

# Load movies
movies = pd.read_csv('movies.csv')

# Content-based recommendation: Similarity in genres
target_movie = "Toy Story (1995)"
target_genres = movies[movies['title'] == target_movie]['genres'].values[0].split('|')

def get_recommendations(movie_title, n=5):
    target_genres = set(movies[movies['title'] == movie_title]['genres'].values[0].split('|'))
    
    # Calculate Jaccard Similarity for genres
    def calculate_similarity(row_genres):
        row_set = set(row_genres.split('|'))
        intersection = target_genres.intersection(row_set)
        union = target_genres.union(row_set)
        return len(intersection) / len(union)

    movies['similarity'] = movies['genres'].apply(calculate_similarity)
    return movies[movies['title'] != movie_title].sort_values('similarity', ascending=False).head(n)

print(f"Top 5 Recommendations for {target_movie}:")
print(get_recommendations(target_movie)[['title', 'genres']])
```
